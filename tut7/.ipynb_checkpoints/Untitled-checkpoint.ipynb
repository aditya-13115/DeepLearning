```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
## Now load the dataset.
```

### Syntax : (xtrain,ytrain), (xtest,ytest) = tf.keras.datasets.mnist.load_data()

```python
(xtrain,ytrain),(xtest,ytest) = tf.keras.datasets.mnist.load_data()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
len(xtest)
```

```python
len(xtrain)
```

```python
for x in range (5):
    plt.matshow(xtrain[x])
```

```python
ytrain[:5]
```

```python
xtrain=xtrain/255
xtest=xtest/255
```

## For ANN we have to flattern the matrix as it only take 1D vector as input,<br>while CNN can have both 2D(black and white) & 3D(for colour images) vectors as input.

```python
xtrain_flat = xtrain.reshape(len(xtrain),28*28)
xtest_flat =xtest.reshape(len(xtest),28*28)
```

```python
xtrain_flat.shape
```

```python
xtest_flat.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation = 'sigmoid')])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(xtrain_flat,ytrain,epochs=5)
```

```python
model.evaluate(xtest_flat,ytest)
```

```python
ypred = model.predict(xtest_flat)
```

```python
ypred[0]
```

```python
for x in range (5):
    print(np.argmax(ypred[x]))
```

```python
ytest[:5]
```

```python
ypred_label = [np.argmax(x) for x in ypred]
```

```python
ypred_label[:5]
```

```python
cm = tf.math.confusion_matrix(labels=ytest, predictions=ypred_label)
```

```python
cm
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True,fmt='d')
plt.ylabel('True')
plt.xlabel('Predicted')
```

## now we will add hidden layers

```python
model1 = keras.Sequential([
    keras.layers.Dense(784,input_shape=(784,),activation = 'relu'),
    keras.layers.Dense(25,activation = 'relu'),
    keras.layers.Dense(10,activation = 'sigmoid')
])

model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model1.fit(xtrain_flat,ytrain,epochs=5)
```

```python
y2pred = model1.predict(xtest_flat)
```

```python
y2pred_label = [np.argmax(x) for x in y2pred]
```

```python
cm2 = tf.math.confusion_matrix(labels=ytest,predictions=y2pred_label)
cm2
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cm2, annot=True,fmt='d')
plt.ylabel('True')
plt.xlabel('Predicted')
```
